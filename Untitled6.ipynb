```python
df_mnist = pd.read_csv(os.path.join(root_path, 'mnist.csv'))
# df_mnist = pd.read_csv("/content/data/mnist.csv")
df_mnist.head()
```

```python
df_2d = pd.read_csv(os.path.join(root_path, 'data_2d.csv'))
# df_2d = pd.read_csv("/content/data/data_2d.csv")
df_2d.head()
```

```python
from sklearn.cluster import KMeans
```

```python
X_2d = df_2d.values
# elbow_range = range(1, 10)
# distortions = []
# for k in elbow_range:
#     km = KMeans(n_clusters=k, random_state=42)
#     km.fit(X_2d)
#     distortions.append(km.inertia_)

def elbow_method(X_2d, max_clusters=10):
    distortions = []
    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X_2d)
        distortions.append(kmeans.inertia_)
    return distortions
```

```python
distortions_2d = elbow_method(X_2d, max_clusters=10)
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), distortions_2d, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel("Inertia (Distortion)")
plt.title("Elbow Method")
plt.show()

```

```python

optimal_clusters_2d = 3

kmeans_2d = KMeans(n_clusters=optimal_clusters_2d, random_state=42)
y_pred_2d = kmeans_2d.fit_predict(X_2d)

plt.figure(figsize=(8, 5))
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y_pred_2d, cmap='viridis')
plt.title(f"K-Means Clustering with  Clusters")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```

```python
x_min, x_max = X_2d[:, 0].min() - 1, X_2d[:, 0].max() + 1
y_min, y_max = X_2d[:, 1].min() - 1, X_2d[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

grid_pred = kmeans_2d.predict(np.c_[xx.ravel(), yy.ravel(), np.zeros_like(xx.ravel())])
grid_pred = grid_pred.reshape(xx.shape)
```

```python
from matplotlib.colors import ListedColormap
import seaborn as sns
# plot the decision boundaries
plt.pcolormesh(xx, yy, grid_pred, cmap=ListedColormap(["r", "g", "b"]), alpha=0.3)

# plot the training points
sns.scatterplot(x=X_2d[:, 0], y=X_2d[:, 1], hue=y_pred_2d)

centroids_x = kmeans_2d.cluster_centers_[:, 0]
centroids_y = kmeans_2d.cluster_centers_[:, 1]
plt.scatter(centroids_x, centroids_y, marker="*", s=60, color="red")
plt.title("Decision ")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
labels = df_mnist.iloc[:, 0]  # Перший стовпець -  мітки
X_mnist = df_mnist.iloc[:, 1:]
X_mnist_scaled = scaler.fit_transform(X_mnist)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
covariance_matrix = np.cov(X_mnist_scaled.T)

# print(covariance_matrix)
```

```python
pca = PCA(n_components =2)
X_mnist_pca = pca.fit_transform(X_mnist_scaled)
X_mnist_pca.shape
# print(X_mnist_pca)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)
# eigenvalues
# eigenvectors
```

```python
#  Сортування власних значень та власних векторів
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

#  Вибір головних компонент
n_components = 2
eigenvectors_subset = sorted_eigenvectors[:, :n_components]

#  Проектування даних на підпростір головних компонент
X_reduced = np.dot(X_mnist_scaled, eigenvectors_subset)
# eigenvectors[:, sorted_indices]
# eigenvectors_subset
```

```python

plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels, cmap='viridis')
plt.colorbar()
plt.title('Проекція  перші дві головні компоненти')
plt.show()
```
